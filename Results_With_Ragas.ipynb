```python
from ragas.dataset_schema import SingleTurnSample
from ragas.metrics._string import NonLLMStringSimilarity
from ragas.metrics import RougeScore, StringPresence
```

```python
sample = SingleTurnSample(
    response="Tom Ridge was born in Munhall, Pennsylvania.",
    reference="Pennsylvania"
)
```

```python
scorer = NonLLMStringSimilarity()
await scorer.single_turn_ascore(sample)
```

```python
scorer = RougeScore()
await scorer.single_turn_ascore(sample)
```

```python
scorer = StringPresence()
await scorer.single_turn_ascore(sample)
```

```python
from langchain_openai import ChatOpenAI
from ragas.llms.base import LangchainLLMWrapper
```

```python
import os

# Prompt the user for their OpenAI API key
api_key = input("Please enter your OpenAI API key: ")

# Set the API key as an environment variable
os.environ["OPENAI_API_KEY"] = api_key

# Optionally, check that the environment variable was set correctly
print("OPENAI_API_KEY has been set!")
```

```python
openai_model = LangchainLLMWrapper(ChatOpenAI(model_name="gpt-4o-mini"))
```

```python
from ragas.dataset_schema import SingleTurnSample
from ragas.metrics._factual_correctness import FactualCorrectness
```

```python
sample = SingleTurnSample(
    response="As of July 1, 2024, the parent company of Reprise Records, the current record label of Stevie Nicks, is Warner Music Group.",
    reference="Warner Music Group"
)
```

```python
scorer = FactualCorrectness()
scorer.llm = openai_model
await scorer.single_turn_ascore(sample)
```

```python
scorer = FactualCorrectness(mode="recall")
scorer.llm = openai_model
await scorer.single_turn_ascore(sample)
```

```python
from datasets import Dataset
from ragas.metrics import answer_correctness
from ragas import evaluate
```

```python
results = []
```

```python
data_samples = {
    'question': {prompt},
    'answer': {response},
    'ground_truth': {answer}
}
dataset = Dataset.from_dict(data_samples)
score = evaluate(dataset,metrics=[answer_correctness])
score.to_pandas()
```

```python
for i in range(30):
    # Get the prompt, response, and answer for the current row
    prompt = df.loc[i, 'Prompt']
    response = df.loc[i, 'model_response']
    answer = df.loc[i, 'Answer']

    # Create a Dataset object for the current row
    data_samples = {
        'question': [prompt],  # Use list for single element
        'answer': [response],  # Use list for single element
        'ground_truth': [answer]  # Use list for single element
    }
    dataset = Dataset.from_dict(data_samples)

    # Calculate the score using `evaluate`
    score = evaluate(dataset, metrics=[answer_correctness])

    # Append the result to the list
    #  (convert to a dictionary for easier DataFrame creation later)
    results.append(score.to_pandas().to_dict('records')[0])

# Create a DataFrame from the results list
final_results_df = pd.DataFrame(results)

# Display the final DataFrame
final_results_df
```

```python
final_results_df.to_csv('ragas_cosinus_score.csv', index=False)
```
